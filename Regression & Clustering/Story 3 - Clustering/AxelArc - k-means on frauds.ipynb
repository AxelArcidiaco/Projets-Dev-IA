```python
# Importation bibliothèques
import pandas as pd

# Importation dataset
data = pd.read_csv("../../Classification/Projet Classification/data/card_credit_fraud_Classification project Final.csv")
data
```

```python
code = {
    'PAYMENT':1,
    'TRANSFER':2,
    'CASH_OUT':3,
    'DEBIT':4,
    'CASH_IN':5,
}

data['type'] = data['type'].map(code)
data = data.drop(['nameOrig', 'nameDest'], axis=1)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split

X = data.drop(columns=['isFraud'], axis = 1)
y = data['isFraud']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.42, random_state=42)
```
