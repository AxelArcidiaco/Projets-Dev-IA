```python
def suite_recur(n):
    res = 1
    if (n == 0):
        return res
    else:
        for i in range(n):
            res = -2*res + 4
        return res
    
# print(suite_recur(0))
# print(suite_recur(5))
# print(suite_recur(13))

nb = 100
for i in range(nb + 1):
    print("v{} = {}".format(i, suite_recur(i)))
```

```python
def suite_un(u0, n):
    res = u0
    if (n == 0):
        return res
    else:
        for i in range(n):
            res = (pow(res, 2) + 1) / (res - 1)
        return res
    
# print(suite_un(5, 10))

for i in range(2, 11):
    for j in range(11):
        print("for u0 = {}, u{} = {}".format(i, j, suite_un(i, j)))
    print("")
```

```python
def calcul_suite_(x0, alpha, elm_num):
    res = x0
    if (elm_num == 0):
        print("x0 = {}".format(res))
    else:
        for i in range(elm_num + 1):
            res = res - alpha * 2 * res - 1
            print("x{} = {}".format(i, res))
        
calcul_suite_(5, 0.3, 25)
```
