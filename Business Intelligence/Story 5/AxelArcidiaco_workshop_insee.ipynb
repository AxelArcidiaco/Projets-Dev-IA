```python
# Importation des bibliothèques et modules Python
import pandas as pd
import seaborn as sns

# Chemin vers les fichiers excel
folder_path = "C:\\Users\\AxelArcidiaco\\Documents\\GRETA\\Projet Dev\\Business Intelligence\\Story 5\\data\\"
path_vaxxed = folder_path + "Part de la population fr vaccinée contre le Covid-19 en 2021.xlsx"
path_screened = folder_path + "Part de la population fr dépistée contre le Covid-19 2020-2021.xlsx"
path_nb_screened_daily = folder_path + "Nombre de dépistage par jour en million.xlsx"
path_nb_hospitalization = folder_path + "Nombre d’hospitalisations pour Covid-19 rapporté à la population régionale pour chaque vague.xlsx"
path_recap = folder_path + "Nombre hebdomadaire hospitalisations -entrées en soins critiques - décès à l'hôpital liés à la Covid-19.xlsx"


# Importation des données
vaccinated_pop = pd.read_excel(path_vaxxed)
screened_pop = pd.read_excel(path_screened)
nb_screened_daily = pd.read_excel(path_nb_screened_daily)
nb_hospitalization = pd.read_excel(path_nb_hospitalization)
recap = pd.read_excel(path_recap)


# Vérification de la réussite de l'importation des données
# vaccinated_pop.head()
# screened_pop.head()
# nb_screened_daily.head()
# nb_hospitalization.head()
# recap.head()
```

# Manipulation de ces données avec Seaborn

```python
# Graphique représentant le nombre d'hospitalisation au début de chaque semaine due au Covid-19 entre le 02/03/2020 et le 06/09/2021
sns.relplot(
    data=recap,
    x="Jour du début de la semaine",
    y="Nouvelles hospitalisations",
    kind="line"
)
```

```python
# Graphique représentant le nombre de nouvelles entrées en soins critiques au début de chaque semaine due au Covid-19
# entre le 02/03/2020 et le 06/09/2021
sns.relplot(
    data=recap,
    x="Jour du début de la semaine",
    y="Nouvelles entrées en soins critiques",
    kind="line"
)
```

```python
# Graphique représentant le nombre de décès au début de chaque semaine due au Covid-19 entre le 02/03/2020 et le 06/09/2021
sns.relplot(
    data=recap,
    x="Jour du début de la semaine",
    y="Décès",
    kind="line"
)
```
