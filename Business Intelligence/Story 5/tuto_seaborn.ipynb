```python
titanic = sbn.load_dataset('titanic')
titanic.head()
```

```python
titanic.drop(['alone','who', 'alive','adult_male','embark_town', 'class'], axis=1, inplace=True)
titanic.dropna(axis=0, inplace=True)
titanic.head()
```

```python
sbn.pairplot(titanic)
```

```python
sbn.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sbn.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sbn.displot(titanic['fare'], kde=True)
```

```python
sbn.jointplot(x='age', y='fare', data=titanic, kind='kde')
```

```python
sbn.jointplot(x='age', y='fare', data=titanic, kind='hex')
```

```python
sbn.heatmap(titanic.corr())
```
